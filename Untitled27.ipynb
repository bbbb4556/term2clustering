```python
import numpy
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# وارد کردن دیتاست
alireza= pd.read_csv('/content/drive/MyDrive/data1.csv')
```

```python
alireza.head()
```

```python
#   آماده سازی دیتاست برای خوشه بندی و حذف ستون لیبل
alireza.drop('label', axis=1, inplace=True)
```

```python
alireza.head()
```

```python
# مشخص کردن تعداد خوشه های لازم
inertia_list=[]
for k in numpy.arange(1, 10):
    kmn= KMeans(n_clusters=k)
    kmn.fit(alireza.values)
    inertia_list.append(kmn.inertia_)
    
inertia_list
```

```python
plt.plot(numpy.arange(1, 10), inertia_list, 'ro-')
plt.xlabel('number of clusters')
plt.ylabel('overlap')
plt.show()
```

```python
# انجام عملیات خوشه بندی
kmn = KMeans(n_clusters=2)
kmn.fit(alireza.values)
labels = kmn.predict(alireza.values)
labels
```

```python
# خوشه بندی صورت گرفته در خروجی کد بالا را با این کد به دیتاست خود متصل می کنیم ومشخص می شود هر ردیف داده در کدام دسته خوشه بندی قرار می گیرد
columns = ['Cluster']
alireza2=pd.DataFrame(labels, index= None, columns=columns)
alireza2.head()
```

```python
# ساخت دیتاست نهایی همراه با لیبل ایجاد شده به نام کلاستر که نشان می دهد هر ردیف از داده جز کدام دسته از خوشه بندی صورت گرفته می باشد
alireza_final=alireza.join(alireza2)
alireza_final
```
